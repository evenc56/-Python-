# 了解一二月民眾借閱偏好—最喜歡什麼樣的作品？

```python
#讀取新北圖書館每月借閱排行版數據:一月
lib_bor_jan = pd.read_excel("https://www.library.ntpc.gov.tw/MainPortalAdmin/document/htmlCnt/2021030217530117-11001%E7%86%B1%E9%96%80%E5%80%9F%E9%96%B1%E6%8E%92%E8%A1%8C%E6%A6%9C.xlsx", header = 2)
```

```python
#檢視資料樣子
lib_bor_jan.head()
```

```python
#下載二月資料後，合併一二月資料
lib_bor_feb = pd.read_excel("https://www.library.ntpc.gov.tw/MainPortalAdmin/document/htmlCnt/2021032417580927-11002%E7%86%B1%E9%96%80%E5%80%9F%E9%96%B1%E6%8E%92%E8%A1%8C%E6%A6%9C.xlsx", header = 2)
lib_bor_all = pd.concat((lib_bor_jan[:-1], lib_bor_feb[:-1]), keys = ('一月', '二月'))
lib_bor_all.shape
```

```python
#最紅的書依序是？
lib_bk_rank = lib_bor_all.groupby('ISBN').agg({'書名':'first', '作者': 'first', '借閱次數':sum}).sort_values('借閱次數', ascending = False)
lib_bk_rank
```

```python
#前五名書籍
a = lib_bk_rank['借閱次數']/sum(lib_bk_rank['借閱次數'])*100
lib_bk_rank = lib_bk_rank.assign( 借閱占比 = a.round(2))
print(lib_bk_rank.shape)
lib_bk_rank[0:5]
```

```python
#前五名作者？
lib_aut_rank = lib_bor_all.groupby('作者').agg({'書名':'first', 'ISBN': 'first', '借閱次數':sum}).sort_values('借閱次數', ascending = False)
lib_aut_rank
b = lib_aut_rank['借閱次數'] / sum(lib_aut_rank['借閱次數']) *100
lib_aut_rank = lib_aut_rank.assign( 借閱占比 = b.round(2))
print(lib_aut_rank.shape)
lib_aut_rank.iloc[0:6, [3]]
```
